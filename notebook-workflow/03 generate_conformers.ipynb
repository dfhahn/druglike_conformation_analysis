```python
import os
from openforcefield.topology import Molecule
from openforcefield.utils.toolkits import ToolkitRegistry
import glob
from simtk import unit
from copy import deepcopy
from tqdm.notebook import tqdm
```

```python
from openforcefield.utils.toolkits import RDKitToolkitWrapper
toolkit_registry = ToolkitRegistry(toolkit_precedence=[RDKitToolkitWrapper])

```

```python
def reconstruct_mol_from_conformer(
    mol: Molecule,
    positions: unit.Quantity,
    identifier: str
) -> Molecule:
    '''
    Adapted from openff-cli/openff/cli/generate_conformers.py
    '''
    mol = deepcopy(mol)
    mol._conformers = None
    mol.add_conformer(positions)
    mol.name = identifier
    return mol
```

```python
os.makedirs('sdf_generated', exist_ok=True)

for file in tqdm(glob.glob(f'sdf_crystal/PDB_*_99.sdf')):
    mol = Molecule.from_file(file, 'SDF', allow_undefined_stereo=True)
    mol.generate_conformers(toolkit_registry=toolkit_registry,
                           n_conformers=10,
                           clear_existing=True)
    mol_name = '_'.join(mol.name.split('_')[:-1])
    for i, conformer in enumerate(mol.conformers):
        identifier = f'{mol_name}_{i:02d}'
        imol = reconstruct_mol_from_conformer(mol, conformer, identifier)
        imol.to_file(f'sdf_generated/{identifier}.sdf', 'SDF')
```
